## x) Top 10 Genres with the Most Content on Netflix

```python
#Extracting unique genres from the 'listed_in' column
all_genres = ', '.join(data['listed_in']).split(', ')

#Get the unique genres by converting the list into a set (removes duplicates) and back into a list
unique_genres = list(set(all_genres))

#Count how many times each genre appears in the list
genre_counts = {}  # Start with an empty dictionary
for genre in unique_genres: #it will iterate the genres one by one(Action,Comedy,Action,Drama,Action,Comedy)
    genre_counts[genre] = all_genres.count(genre)  # it will Count each genre's occurrences in all_gnres list

#Get the top 10 genres with the most content
top_genres = dict(sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)[:10])

#genre_counts.items() => This takes the genre_counts dictionary and converts it into a list of key-value pairs (tuples). example:[('Action', 3), ('Comedy', 2), ('Drama', 1)]
#reverse=True => it would sort in ascending order (lowest count to highest count).
```

```python
#Create a horizontal bar plot using Matplotlib
plt.barh(list(top_genres.keys()), list(top_genres.values()), color='purple')

#Add labels and title
plt.xlabel('Number of Contents')
plt.ylabel('Genre')
plt.title('Top 10 Genres with the Most Content on Netflix')

# Show the plot
plt.show()
```
